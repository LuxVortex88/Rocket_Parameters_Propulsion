```python
import matplotlib.pyplot as plt
import numpy as np

# Constants
g0 = 9.81    # gravity at sea level (m/s^2)


# Rocket parameters
m0 = 500 # initial mass (kg)
mf = 100 # final mass (kg)
ve = 2500 # effective exhaust velocity (m/s)

# Tsiolkovsky rocket equation
delta_v = ve * np.log(m0 / mf)

print(f"Delta V: {delta_v:.2f} m/s")

# Simulate velocity over mass loss
mass = np.linspace(m0, mf, 100)
velocity = ve * np.log(m0 / mass)

plt.plot(mass, velocity)
plt.xlabel('mass (kg)')
plt.ylabel('velocity (m/s)')
plt.title('Rocket velocity vs Mass')
plt.grid()
plt.show()
```
